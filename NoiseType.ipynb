## Noise Type

This notebook contains all the types of noises that we have applied to the images.

```python
import os

from debruiteur.plots.plots import plot_im_grid_from_df
from debruiteur.preprocessing.preprocessor import make_original_dataframe, make_resized_dataframe
from debruiteur.noise.noise import GaussianNoise, PoissonNoise, UniformNoise, SaltPepperNoise, SquareMaskNoise, SpeckleNoise
```

### Load data

```python
working_dir = os.path.abspath(os.getcwd())

df_original = make_original_dataframe(os.path.join(working_dir, "images"))
df_resized = make_resized_dataframe(df_original, img_shape=(100, 100), resized_path=os.path.join(working_dir, "resized_images"))
```

### Noise

plot_im_grid_from_df(df, noise, rows, columns, figsize)

Parameters :
* df : dataframe, with first column containing image's path
* noise : Sublclass of Noise
* rows : Number of rows in image grid
* columns : Number of columns in image grid
* figsize : Tuple, shape of subfigure

#### None

```python
plot_im_grid_from_df(df_resized)
```

#### Gaussian noise

A gaussian noise is statistical noise having a probability density function equal to that of the gaussian distribution. 

Principal sources of Gaussian noise in digital images arise during acquisition e.g. sensor noise caused by poor illumination and/or high temperature, and/or transmission e.g. electronic circuit noise.

Parameters :
* mean
* std

```python
mu, sigma = 0, 40
plot_im_grid_from_df(df_resized, GaussianNoise(mean=mu, std=sigma))
```

#### Poisson noise

Poisson noise is a type of noise which can be modeled by a Poisson process. In electronics poisson noise originates from the discrete nature of electric charge. Poisson noise also occurs in photon counting in optical devices, where shot noise is associated with the particle nature of light.

Parameters : 

* factor

```python
factor = 3
plot_im_grid_from_df(df_resized, PoissonNoise(factor=factor))
```

#### Uniform noise

Uniform noise results from the quantization of the pixels in an image. It has a uniform distribution but can be signal-dependent.

Parameters :
* ampltitude : in ]0; 255[

```python
plot_im_grid_from_df(df_resized, UniformNoise(100))
```

#### Salt and pepper noise

This noise can be caused by sharp and sudden disturbances in the image signal. It presents itself as sparsely occurring white and black pixels.

Parameters :
* p : probability in ]0;1[

```python
plot_im_grid_from_df(df_resized, SaltPepperNoise(0.05))
```

#### Squared mask

This type of noise may be caused by a sensor failure or a stain on the lens. This type of noise is characterized by missing parts of the image.

Parameters :
* mask_shape : tuple, (width, height)
* target_freq : frequence in ]0; 1[

```python
plot_im_grid_from_df(df_resized, SquareMaskNoise(mask_shape=(10, 10), freq=0.1))
```

#### Speckle

Speckle is a form of multiplicative noise, which occurs when a pulse of a sound wave arbitrarily interferes with small particles or objects on a scale comparable to the wavelength of the sound. 

No parameters

```python
plot_im_grid_from_df(df_resized, SpeckleNoise())
```
